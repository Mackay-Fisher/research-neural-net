```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,1,100)
y = np.sqrt(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
X_train = X_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)
```

```python
model = keras.Sequential([
    keras.layers.Dense(140, activation = 'relu', kernel_initializer = 'random_normal', input_shape = (1,)),
    keras.layers.Dense(1)  
])
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error' )
```

```python
e = 1000
model.fit(X_train, y_train, epochs=e)
```

```python
X_train.shape
```

```python
val = model.predict(X_test)
```

```python
plt.scatter(X_train, y_train)
plt.scatter(X_train, model.predict(X_train))
sorted_data = np.array(sorted(X_train, key=lambda x: x[0]))
print(sorted_data)
plt.plot(sorted_data, model.predict(sorted_data), color='red')
```
